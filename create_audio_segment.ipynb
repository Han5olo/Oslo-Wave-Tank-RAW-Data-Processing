```python

custom_audio_path = None
custom_csv_path = '/home/dorian/Desktop/OsloCSV/rec'

# %%
import scipy.io.wavfile as wav
from pathlib import Path
```

```python
csv_path = Path(custom_csv_path or Path.cwd() / 'csv')
csv_files = [f for f in csv_path.glob('*.csv')]

audio_path = Path(custom_audio_path or Path.cwd() / 'demo')
audio_files = sorted([f for f in audio_path.rglob('*.WAV')])

audio_file = audio_files[0]  # For testing, use the first audio file

#%% Read audio file
samplerate, data = wav.read(audio_file)
```

```python
start = 210  # seconds
end = 216    # seconds
```

```python
start_sample = int(start * samplerate)
end_sample = int(end * samplerate)

segment = data[start_sample:end_sample]

wav.write("segment.wav", samplerate, segment)
```
